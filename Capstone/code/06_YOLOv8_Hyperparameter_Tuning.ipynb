```python
import subprocess

# Hyperparameter grid
learning_rates = [0.001, 0.01, 0.1]
batch_sizes = [8, 16]

# Path to your data and model
data_path = "/home/mangguai/capstone/code/Master_Dataset/data.yaml"
model_path = "yolov8s.pt"
epochs = 10
imgsz = 640

# Loop through hyperparameter combinations
for lr in learning_rates:
    for batch in batch_sizes:
        command = f"yolo train model={model_path} data={data_path} epochs={epochs} imgsz={imgsz} batch={batch} lr0={lr}"
        subprocess.run(command, shell=True)

# 391 minutes for 6 combinations
```
